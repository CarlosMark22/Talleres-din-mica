Calcule:

<h1> Resolución </h1> 
Primero se realiza un análisis de fuerzas en el bloque.



Como el bloque se encuentra rodando sin deslizar, entonces se cumple la Primera Ley de Newton. Entonces, realizando un sumatorio de fuerzas en y se tiene:<br>
$\sum{\vec{F_y}}=0$<br>
$\vec{N_y}-\vec{P}+\vec{f_{remax}}=0$<br>
$Ncos(\theta)-mg+\mu_eNsen(\theta)=0$<br>
$N(cos(\theta)+\mu_esen(\theta))=mg$<br>
(1)$N=\frac{mg}{cos(\theta)+\mu_esen(\theta)}$<br>
Luego, realizando un sumatorio de fuerzas en x:<br>
$\sum\vec{F_x}=0$<br>
$-\vec{f_{remax}}+\vec{N_x}=0$<br>
(2)$N(sen(\theta)-\mu_e(cos\theta))=0$<br>
Resolviendo el sistema de ecuaciones para (1) y (2) se tiene: <br>
$\frac{mg(sen(\theta)-\mu_ecos(\theta))}{cos(\theta)+\mu_esen(\theta)}=0$<br>
$sen(\theta)-\mu_ecos(\theta)=0$<br>
$tan(\theta)=\mu_e$<br>
$\theta=16.699$<br>
Luego, el gráfico de la normal con respecto al ángulo viene dado por:<br>

```python
# **Gráfico** 
import numpy as np #importar libreria matematica
import plotly.express as px #herramienta de plot
import plotly.graph_objects as go

t = np.linspace(0,16,699)
N =(2000*9.8)/(np.cos(np.radians(t))+0.3*(np.sin(np.radians(t))))  
fig=px.line(x=t,y=N,labels={'x':'Ángulo [°]','y':'Normal[N]'},title="Fuerza normal vs tiempo del bloque")
fig.show()
```
